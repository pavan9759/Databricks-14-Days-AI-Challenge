```python
%sql
CREATE SCHEMA IF NOT EXISTS ecommerce.analytics;

CREATE TABLE IF NOT EXISTS ecommerce.analytics.sales (
    order_id STRING,
    order_date DATE,
    customer_id STRING,
    product STRING,
    category STRING,
    quantity INT,
    price DOUBLE,
    revenue DOUBLE
);

INSERT INTO ecommerce.analytics.sales VALUES
('O001','2025-01-01','C001','iPhone','Mobile',1,999,999),
('O002','2025-01-01','C002','Samsung','Mobile',1,799,799),
('O003','2025-01-02','C003','MacBook','Laptop',1,1299,1299),
('O004','2025-01-02','C001','AirPods','Accessories',2,199,398),
('O005','2025-01-03','C004','iPhone','Mobile',1,999,999),
('O006','2025-01-03','C005','Dell XPS','Laptop',1,1199,1199),
('O007','2025-01-04','C002','Mouse','Accessories',3,49,147),
('O008','2025-01-04','C003','Keyboard','Accessories',2,99,198),
('O009','2025-01-05','C006','Samsung','Mobile',1,799,799),
('O010','2025-01-05','C007','MacBook','Laptop',1,1299,1299);

SHOW TABLES IN ecommerce.analytics;


--Validate Data
select * from ecommerce.analytics.sales;

-- Analytical Queries (Day 9)
-- Total Revenue (KPI)
SELECT SUM(revenue) AS total_revenue
FROM ecommerce.analytics.sales;


--Revenue Trend (Date-wise)
SELECT
    order_date,
    SUM(revenue) AS daily_revenue
FROM ecommerce.analytics.sales
GROUP BY order_date
ORDER BY order_date;


--Revenue by Category
SELECT
    category,
    SUM(revenue) AS category_revenue
FROM ecommerce.analytics.sales
GROUP BY category
ORDER BY category_revenue DESC;

--Top Product
SELECT
    product,
    SUM(quantity) AS total_units,
    SUM(revenue) AS revenue
FROM ecommerce.analytics.sales
GROUP BY product
ORDER BY revenue DESC
LIMIT 5;

-- 
```
